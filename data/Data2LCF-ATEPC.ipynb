```python
review = list()
aspects_review = list()
polarity_review = list()

for group in dataDf.groupby("id"):
    group_content = group[1]
    print("======================================= Review {}".format(group_content.iloc[0]['id']))
    
    r = group_content.iloc[0]['review'].split()
    print(group_content.iloc[0]['review'])
    l_r = len(r)
    aspects = list()
    polarity = list()
    for sentence_group in group_content.groupby('index_sentence'):
        print("========================== Sentence")
        current_sentence = sentence_group[1].iloc[0]['sentence']
        print(current_sentence)
        current_sentence_splitted = current_sentence.split()
        
        
        
        index_words = [0 for i in range(len(current_sentence_splitted))]
        
        for i in range(1, len(current_sentence_splitted)):
            index_words[i] = len(current_sentence_splitted[i-1])+1+index_words[i-1]
            
        index_words += [len(current_sentence)]
        
        #print(current_sentence_splitted)
        #print(index_words)
        
        aspects_sentence = [0 for i in range(len(current_sentence_splitted))]
        polarity_sentence = [-1 for i in range(len(current_sentence_splitted))]
        
        index_words_to_label = []
        
        for index, sentence in sentence_group[1].iterrows():
            if(int(sentence['from']) in index_words):
                if(sentence['from'] != 0 and sentence['to'] != 0):
                    print("============= Aspect")
                    print(sentence['target'], sentence['from'], sentence['to'])
                    ind = current_sentence.split()

                    index_first_word = index_words.index(int(sentence['from']))
                    val = index_words[index_first_word+1]
                    aspects_sentence[index_first_word] = "B-ASP"

                    if(sentence['polarity'] == "negative"):
                        polarity_sentence[index_first_word] = 0
                    elif(sentence['polarity'] == "positive"):
                        polarity_sentence[index_first_word] = 2

                    while val < int(sentence['to']):
                        index_first_word += 1
                        aspects_sentence[index_first_word] = "I-ASP"

                        if(sentence['polarity'] == "negative"):
                            polarity_sentence[index_first_word] = 0
                        elif(sentence['polarity'] == "positive"):
                            polarity_sentence[index_first_word] = 2

                        val = index_words[index_first_word+1]
                    
        print("============= Result")
        
        print(aspects_sentence)
        print(polarity_sentence)
        aspects += aspects_sentence
        polarity += polarity_sentence
        
    print(l_r)
    print(len(aspects))
    print(len(polarity))   
    review.append(r)
    aspects_review.append(aspects)
    polarity_review.append(polarity)
    
print(len(review))
print(len(aspects_review))
print(len(polarity_review))     
```

```python
with open("ter.atepc.train.dat", "a+") as file_object:
    for i in range(len(review)):
        for word in range(len(review[i])):
            print(len(review[i]))
            print(len(aspects_review[i]))
            print(len(polarity_review[i]))
            # Open the file in append & read mode ('a+')

            # Move read cursor to the start of file.
            file_object.seek(0)
                # If file is not empty then append '\n'
            data = file_object.read(100)
            if len(data) > 0 :
                file_object.write("\n")
                # Append text at the end of file
            file_object.write("{} {} {}".format(review[i][word], aspects_review[i][word], polarity_review[i][word]))
        file_object.write("\n")
```
